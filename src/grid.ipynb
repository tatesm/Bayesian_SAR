```python
# Start by creating a 20x20 grid of zeros, will then fill in with gaussian like wide probablities, 10x10 being the center

import numpy as np
import math

grid_arr = np.zeros((20,20))
```

```python
cx = 10
cy = 10

sigma = 4


for (i, j), value in np.ndenumerate(grid_arr):
    distance = math.sqrt((i-cx)**2+(j-cy)**2)
    score = np.exp((-distance**2)/(2*(sigma**2)))
    grid_arr[i][j] = score
    print(f"Index: ({i}, {j}), Distance: {distance}, Score: {score}")

```

```python
# normalize each score in each cell

grid = grid_arr /np.sum(grid_arr)
np.sum(grid)
```

```python
# Visualize grid

for i in range(grid.shape[0]):
    for j in range(grid.shape[1]):
        print(f"{grid[i,j]:.3f}", end=" ")
    print()
```

```python
# write a function to update grid after search. If searched an not found, searched grid prob goes down, normalize entire grid, results in other cells probs increasing.

def grid_update(cell, grid, pod):
    i, j = cell
    new_grid = grid.copy()
    new_grid[i,j] = new_grid[i,j] * (1-pod)
    new_sum = np.sum(new_grid)
    updated_grid = new_grid/new_sum
    if np.isclose(np.sum(updated_grid), 1):
        return updated_grid
    else:
        raise ValueError(f"Updated grid total probability != 1, total probability is {np.sum(updated_grid)}")
```

```python
## Basic test

cell = (10,10)
pod = 0.5

updated_grid = grid_update(cell, grid, pod)

for i in range(grid.shape[0]):
    for j in range(grid.shape[1]):
        print(f"{updated_grid[i,j]:.3f}", end=" ")
    print()
```

```python
# Test for cell itself and for neighboring cell

print(f"Before update: {grid[10,10]}, post update: {updated_grid[10,10]}")

print(f"Neighbor before: {grid[10,11]}, and after: {updated_grid[10,11]}")
```

```python
# Repeated cell reduction test

another_grid = grid_update(cell, updated_grid, pod)

print(f"Before update: {updated_grid[10,10]}, post update: {another_grid[10,10]}")

print(f"Neighbor before: {updated_grid[10,11]}, and after: {another_grid[10,11]}")
```

```python
# in V1 model, POD is constant across entire gird so POS == POA, will add variable POD in next version

def choose_next_cell(current_grid):
    k = np.argmax(current_grid)
    next_cell = np.unravel_index(k,current_grid.shape)
    return next_cell
```
