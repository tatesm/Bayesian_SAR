```python
# init grid function initializes a blank n x m grid

def init_grid(rows, columns):
    grid = np.zeros((rows,columns))
    return grid
```

```python
# function to define what prior probabiliity distribution you want to model

def prior_probability_distribution(distribution_type, **params):
    if distribution_type == "uniform":
        def uniform_function(i,j):
            score = 1
            return score
        return uniform_function

    elif distribution_type == "gaussian":
        # extract params
        cx = params["cx"]
        cy = params["cy"]
        sigma = params["sigma"]
        # define gaussian_function
        def gaussian_function(i,j):

            distance = math.sqrt((i-cx)**2+(j-cy)**2)
            score = np.exp((-distance**2)/(2*(sigma**2)))
            return score

        return gaussian_function

    else:
        raise ValueError("Unsupported distribution choice")
    
```

```python
# function that fills out grid with chosen probability
def apply_prior_distribution(grid, cell_probability_function):
    new_grid = np.zeros_like(grid)
    for (i, j), value in np.ndenumerate(grid):
        new_grid[i, j] = cell_probability_function(i,j)
        print(f"Index: ({i}, {j}), Cell Probability: {new_grid[i, j]}")
    return new_grid
    
```

```python
def normalize_grid(grid):
    if np.isclose(np.sum(grid), 0):
        raise ValueError("Sum of filled grid is 0")
    normalized_grid = grid /np.sum(grid)
    if np.isclose(np.sum(normalized_grid),1):
        return normalized_grid
    else:
        raise ValueError(f"Normalized grid total probability != 1, total probability is {np.sum(normalized_grid)}")
```

```python
# write a function to update grid after search. If searched an not found, searched grid prob goes down, normalize entire grid, results in other cells probs increasing.

def grid_update(cell: tuple, grid: object, pod: float) -> object:
    i, j = cell
    new_grid = grid.copy()
    new_grid[i,j] = new_grid[i,j] * (1-pod)
    new_sum = np.sum(new_grid)
    updated_grid = new_grid/new_sum
    if np.isclose(np.sum(updated_grid), 1):
        return updated_grid
    else:
        raise ValueError(f"Updated grid total probability != 1, total probability is {np.sum(updated_grid)}")
```

```python
# in V1 model, POD is constant across entire gird so POS == POA, will add variable POD in next version

def choose_next_cell(current_grid: object):
    k = np.argmax(current_grid)
    next_cell = np.unravel_index(k,current_grid.shape)
    return next_cell
```

```python
def run_search(grid: object, pod: float, steps: int) -> list: 
    search_history = []
    # grid_history = [] will be used in next version, to see how grid shifts and evolves

    for i in range(steps):
        cell = choose_next_cell(grid)
        search_history.append(cell)
        grid = grid_update(cell, grid, pod)
    return search_history
        
        
```

## Visualization Idea: 3D Search Evolution

### Goal
Create an intuitive visualization of how the search algorithm explores the probability grid over time.

---

### Core Concept

Represent the probability grid as a **3D surface**:

- **X, Y axes** → grid coordinates  
- **Z axis** → probability (POA)  
- This creates a “landscape” where:
  - Peaks = high probability regions  
  - Valleys = low probability regions  

---

### Search Visualization

Overlay the **search process** on top of the surface:

#### Option 1: Path Line
- Draw a line connecting searched cells in order  
- Shows the trajectory of the algorithm  
- Helps visualize how the search moves across the probability landscape  

#### Option 2: Cell Highlighting
- Change color of searched cells over time  
- Example:
  - Red = recently searched  
  - Fading to darker tones = older searches  
- Makes search coverage visually clear  

#### Option 3: Animated Steps (Best)
- Animate each step of the search:
  1. Highlight current cell  
  2. Update the surface (probabilities shift)  
  3. Move to next cell  

---

### What This Reveals

This visualization will help show:

- How the algorithm follows **probability gradients**
- Natural **expansion from high-probability regions**
- Differences in behavior based on:
  - POD values  
  - Shape of the prior (Gaussian vs flat vs multi-peak)  
- Whether the algorithm:
  - Explores efficiently  
  - Gets stuck or cycles
